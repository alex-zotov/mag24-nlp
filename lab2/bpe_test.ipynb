# BPE.fit test

тест1

```python
tokenizer = bpe.BPE(28)
tokenizer.fit('На дворе дрова, за двором дрова, дрова вширь двора, не вместит двор дров, надо дрова выдворить на дровяной двор.')
```

```python
tokenizer.id2token.items()
```

```python
'''
' _,_._Н_а_в_д_е_з_и_й_м_н_о_р_с_т_ш_ы_ь_я_ д_ро_во_ дро_ дров_ дво_ двор'
'''
len(tokenizer.id2token),'_'.join(tokenizer.id2token.values())
```

тест2

```python
tokenizer = bpe.BPE(30)
tokenizer.fit('Из кузова в кузов шла перегрузка арбузов. В грозу в грязи от груза арбузов развалился кузов.')
```

```python
tst = "_".join([' ','.','В','И','а','б','в','г','е','з','и','к','л','о','п','р','с','т','у','ш','я','уз','узо','узов','а ','гр',' к',' кузов',' гр','а а'])
print(f'{tst=}')
len(tokenizer.id2token),'_'.join(tokenizer.id2token.values())
```

тест3

```python
tokenizer = bpe.BPE(31)
tokenizer.fit('Однажды был случай в далёком Макао: макака коалу в какао макала, коала лениво какао лакала, макака макала, коала икала.')
```

```python
tst = "_".join([' ',',','.',':','М','О','а','б','в','д','е','ж','и','й','к','л','м','н','о','с','у','ч','ы','ё','ка','ла','ака','ко',' м',' мака',' ко'])
print(f'{tst=}')
len(tokenizer.id2token),'_'.join(tokenizer.id2token.values())
```

```python
print((sum(tokenizer.token2id.values()), '_'.join(tokenizer.id2token.values())))
```
