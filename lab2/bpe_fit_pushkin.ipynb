```python
import bpe
```

# BPE.fit на стихах Пушкина

```python
text = ''
with open('../data/pushkin-onegin.txt') as f:
    text=f.read()

text[:100],len(text)
```

```python
len(set(text)),list(set(text))
```

```python
tokenizer = bpe.BPE(2000)
tokenizer.fit(text)
```

```python
len(tokenizer.id2token),[tokenizer.id2token[key] for key in sorted(tokenizer.id2token.keys())[-20:]]
```

# BPE.save  BPE.load

```python
bpe_file = './models/bpe_onegin.dill'
tokenizer.save(bpe_file)
```

```python
test='Роман в стихах\nНе мысля гордый свет забавить,\nВниманье дружбы возлюбя'
emb = tokenizer.encode(test)
print(emb)
[tokenizer.id2token[id] for id in emb]
```

```python
tokenizer=None
tokenizer=bpe.BPE.load(bpe_file)
```
