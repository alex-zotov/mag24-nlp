```python
import torch
import decoder

torch.manual_seed(10)
```

# Decoder



```python
device = 'cpu'
batch_size = 2
seq_len = 40

max_seq_len = 300
vocab_size = 100
emb_size = 128

head_size = 4
num_heads = 3
dropout = 0.3
```

```python
# на входе тензор (batch_size,seq_len,emb_size)
x = torch.rand(size=(batch_size,seq_len,emb_size),device=device)
x.shape, x.dtype
```

```python
decoder = decoder.Decoder(
    num_heads=num_heads,
    emb_size=emb_size,
    head_size=head_size,
    max_seq_len=max_seq_len,
    dropout=dropout,
    device=device)
```

```python
#  на выходе та же размерность batch_size x seq_len x emb_size
y = decoder.forward(x)
y.shape, y.dtype
```
