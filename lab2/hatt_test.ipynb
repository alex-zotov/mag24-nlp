```python
import torch
import hatt

torch.manual_seed(10)
```

# Head Attention



```python
device = 'cpu'
batch_size = 2
seq_len = 40

max_seq_len = 300
vocab_size = 100
emb_size = 128

head_size = 4
```

```python
# на входе тензор (batch_size,seq_len,emb_size)
x = torch.rand(size=(batch_size,seq_len,emb_size),device=device)
x.shape, x.dtype
```

```python
head_attention = hatt.HeadAttention(emb_size=emb_size,head_size=head_size,max_seq_len=max_seq_len,device=device)
```

```python
#  на выходе batch_size x seq_len x head_size
y = head_attention.forward(x)
y.shape, y.dtype
```
