```python
from torch.utils.data import Subset, DataLoader
import data_loader
import bpe
```

# DataLoader

```python
data_file = '../data/pushkin-onegin.txt'
bpe_file = './models/bpe_onegin.dill'
seq_len=100
device='cpu'
```

```python
text = ''
with open(data_file) as f:
    text=f.read()

text[:100],len(text)
```

```python
tokenizer = bpe.BPE.load(bpe_file)
len(tokenizer.id2token)
```

```python
tokens = tokenizer.encode(text)
len(tokens)
```

```python
dataset = data_loader.GetData(data= tokens, seq_len= 50, device= device)
```

```python
samples = [dataset[i] for i in range(10)]
```
